# Python Programlama Dilinin Temel Özellikleri

Python, geniş bir kullanıcı kitlesine hitap eden, basit ve güçlü bir programlama dilidir. Özellikle başlangıç ​​seviyesindeki öğrencilere ve programcılara uygun olmasıyla bilinir. İşte Python'un temel özellikleri:

## 1. **Basit ve Anlaşılır Sözdizimi:**
Python, okunabilir ve anlaşılır bir sözdizim sunar. Bu sayede, kodu yazmak ve anlamak daha kolay hale gelir. Belirgin bir blok yapısı, indentasyon (girinti) kullanımı sayesinde sağlanır.

Örnek:
```python
if x > 0:
    print("x pozitif bir sayıdır.")
else:
    print("x negatif veya sıfırdır.")
```

## 2. **Yüksek Seviyeli Dil:**
Python, yüksek seviyeli bir programlama dilidir. Bu, kullanıcının karmaşık işlemleri daha az kodla gerçekleştirebileceği anlamına gelir. Özellikle veri manipülasyonu ve veri analizi gibi alanlarda avantaj sağlar.

Örnek:
```python
toplam = sum([1, 2, 3, 4, 5])
print("Toplam:", toplam)
```

## 3. **Modüler ve Geniş Kütüphane Desteği:**
Python'un geniş kütüphane desteği, kullanıcıların birçok farklı görevi kolayca gerçekleştirmesine olanak tanır. Kütüphaneler, hazır fonksiyonlar ve araçlar içerir, böylece tekrarlayan işlemleri yeniden yazmak zorunda kalmazsınız.

Örnek:
```python
import math

kare_kok = math.sqrt(25)
print("Kare Kök:", kare_kok)
```

## 4. **Çeşitli Veri Türleri:**
Python, çeşitli veri türlerini destekler, bu da programcılara esneklik sağlar. Temel veri türleri arasında sayılar, metin (string), liste, demet, küme ve sözlük bulunur.

Örnek:
```python
isim = "Python"
liste = [1, 2, 3, 4, 5]
sozluk = {'anahtar': 'değer'}
```

## 5. **Nesne Yönelimli Programlama (OOP) Desteği:**
Python, nesne yönelimli programlamayı destekler. Bu, kodunuzu daha organize etmenize, sürdürmenize ve tekrar kullanmanıza yardımcı olur. Sınıflar ve nesneler kullanarak programlama yapabilirsiniz.

Örnek:
```python
class Araba:
    def __init__(self, marka, model):
        self.marka = marka
        self.model = model

    def bilgileri_goster(self):
        print(f"{self.marka} {self.model}")

araba1 = Araba("Toyota", "Corolla")
araba1.bilgileri_goster()
```

Bu temel özellikler, Python'un neden popüler bir programlama dil olduğunu gösteriyor. Öğrencilere bu temel kavramları öğretmek, onların programlama dünyasına adım atmalarına yardımcı olacaktır.

```python
class Insan:
  def __init__(self, isim):
    self.isim = isim

  def boy_bilgisi_gir(self,boy):
    self.boy = boy

class Kisi(Insan):
  def sever(self, sevdigi_sey):
    self.sever = sevdigi_sey

```

# Python'da Temel Veri Tipleri

Python, çeşitli veri tiplerini destekleyen esnek bir programlama dilidir. Her veri tipi, belirli türdeki verileri temsil eder ve farklı işlemleri destekler. İşte Python'da kullanılan temel veri tipleri:

## 1. **Integer (Tam Sayı):**
Tam sayılar, negatif veya pozitif olabilen, ondalık kısmı olmayan sayılardır.

Örnek:
```python
sayi = 42
```

## 2. **Float (Ondalık Sayı):**
Ondalık sayılar, ondalık kısmı olan sayılardır.

Örnek:
```python
ondalik_sayi = 3.14
```

## 3. **String (Karakter Dizisi):**
Karakter dizileri, metin verilerini temsil eder. Tek tırnak (' ') veya çift tırnak (" ") içinde tanımlanır.

Örnek:
```python
isim = "Python"
```

## 4. **Boolean (Mantıksal Değer):**
Boolean değerleri, `True` veya `False` olabilir ve genellikle karar yapılarında kullanılır.

Örnek:
```python
durum = True
```

## 5. **List (Liste):**
Listeler, sıralı ve değiştirilebilir veri öğelerini içeren veri tipleridir.

Örnek:
```python
liste = [1, 2, 3, 4, 5]
```

## 6. **Tuple (Demet):**
Demetler, sıralı ancak değiştirilemez veri öğelerini içeren veri tipleridir.

Örnek:
```python
demet = (1, 2, 3)
```

## 7. **Set (Küme):**
Kümeler, benzersiz ve sırasız veri öğelerini içeren veri tipleridir.

Örnek:
```python
kume = {1, 2, 3, 4}
```

## 8. **Dictionary (Sözlük):**
Sözlükler, anahtar-değer çiftlerini içeren veri tipleridir. Her bir anahtar, bir değerle ilişkilidir.

Örnek:
```python
sozluk = {'anahtar1': 'deger1', 'anahtar2': 'deger2'}
```

Bu temel veri tipleri, Python programlamasında çeşitli görevler için kullanılır. Programcılar, bu veri tiplerini bir araya getirerek karmaşık veri yapıları oluşturabilir ve çeşitli işlemleri gerçekleştirebilirler. Python'un esnek veri tipleri, dilin geniş bir uygulama yelpazesi içinde kullanılmasına olanak tanır.

Python'da karşılaştırma ve mantık operatörleri, programcılara koşullar oluşturarak kontrol sağlama imkanı tanır. İşte Python'da kullanılan bazı karşılaştırma ve mantık operatörleri ile örnekler:

## Karşılaştırma Operatörleri:

### 1. Eşitlik (==):

```python
x = 5
y = 7

if x == y:
    print("x, y'ye eşittir.")
else:
    print("x, y'ye eşit değildir.")
```

### 2. Eşit Değil (!=):

```python
a = 10
b = 20

if a != b:
    print("a, b'ye eşit değildir.")
else:
    print("a, b'ye eşittir.")
```

### 3. Büyük veya Eşit (>=), Küçük veya Eşit (<=):

```python
m = 30
n = 25

if m >= n:
    print("m, n'den büyük veya eşittir.")
else:
    print("m, n'den küçüktür.")
```

## Mantık Operatörleri:

### 1. Mantıksal VE (and):

```python
sayi1 = 10
sayi2 = 5

if sayi1 > 0 and sayi2 > 0:
    print("Her iki sayı da pozitiftir.")
else:
    print("En az bir sayı pozitif değildir.")
```

### 2. Mantıksal VEYA (or):

```python
giris = input("Kullanıcı adınızı veya e-postanızı girin: ")

if giris.isdigit() or '@' in giris:
    print("Geçerli bir kullanıcı adı veya e-posta girdiniz.")
else:
    print("Geçerli bir kullanıcı adı veya e-posta giriniz.")
```

### 3. Mantıksal DEĞİL (not):

```python
dogru_mu = True

if not dogru_mu:
    print("Değişken 'True' değil.")
else:
    print("Değişken 'True'.")
```

Bu örneklerde gördüğünüz gibi, karşılaştırma operatörleri **(`==`, `!=`, `>=`, `<=`, vb.)** ile değişkenlerin durumunu kontrol edebilir, mantık operatörleri **(`and`, `or`, `not`)** ile birden fazla koşulu birleştirebilir ve karmaşık kontrol ifadeleri oluşturabilirsiniz. Bu operatörler, programlarınızda mantıklı akış kontrolü sağlamak için kullanılır.

Python'da koşul ifadeleri, belirli bir koşulun doğru (True) veya yanlış (False) olmasına bağlı olarak belirli bir kod bloğunun çalıştırılmasını sağlar. Koşul ifadeleri, "if", "else" ve "elif" (else if) ifadelerini içerir. İşte bu ifadelerle ilgili örnekler:

## `if` İfadesi:

```python
yas = 18

if yas >= 18:
    print("Ehliyet alabilirsiniz.")
```

Bu örnekte, `yas` değişkeninin değeri 18 veya daha büyükse, ekrana "Ehliyet alabilirsiniz." yazdırılır.

## `if` ve `else` İfadeleri:

```python
puan = 75

if puan >= 50:
    print("Geçtiniz!")
else:
    print("Kaldınız.")
```

Bu örnekte, `puan` değişkeninin değeri 50 veya daha büyükse, "Geçtiniz!" yazdırılır; aksi takdirde "Kaldınız." yazdırılır.

## `if`, `elif`, ve `else` İfadeleri:

```python
not_ortalamasi = 75

if not_ortalamasi >= 90:
    print("AA")
elif not_ortalamasi >= 80:
    print("BA")
elif not_ortalamasi >= 70:
    print("BB")
else:
    print("Başarısız.")
```

Bu örnekte, `not_ortalamasi` değişkenine bağlı olarak öğrencinin notunu belirleyen bir if-elif-else yapısı kullanılmıştır. Birden fazla koşulu kontrol eder ve ilk doğru koşulu sağlayan bloğu çalıştırır.

## İç İçe Koşullar:

```python
sayi = 10

if sayi > 0:
    if sayi % 2 == 0:
        print("Bu sayı pozitif ve çift.")
    else:
        print("Bu sayı pozitif ve tek.")
else:
    print("Bu sayı negatif veya sıfır.")
```

Bu örnekte, iç içe koşullar kullanılarak, bir sayının pozitif ve çift mi, pozitif ve tek mi, yoksa negatif veya sıfır mı olduğu kontrol edilmiştir.

Koşul ifadeleri, programlarınızın belirli durumlarda farklı aksiyonlar almasına olanak tanır ve programınızı daha esnek ve dinamik hale getirir.

## Döngüler (Loops):

sayac = 0
while sayac < 5:
    print(sayac)
    sayac += 1

Bu örnekte, sayacın 5'e kadar olan değerleri ekrana yazdıracak bir `while` döngüsü bulunmaktadır.

## Veri Yapıları:

Python'da fonksiyonlar, belirli bir görevi yerine getiren, tekrar kullanılabilir ve modüler kod bloklarıdır. Python'da fonksiyon tanımlama, çağırma ve parametreleri işleme konularında oldukça esnek bir yapı vardır. İşte Python'da fonksiyonlarla ilgili temel kavramları açıklayan bir yazı:

## Fonksiyon Tanımlama:

Python'da bir fonksiyon, `def` anahtar kelimesi ile tanımlanır. Fonksiyonlar, giriş parametreleri alabilir ve bir veya birden fazla çıkış değeri üretebilir.

```python
def merhaba_dunya():
    print("Merhaba, Dünya!")

merhaba_dunya()
```

Bu örnekte, `merhaba_dunya` adlı basit bir fonksiyon tanımlanmış ve çağrılmıştır.

## Parametreler ve Dönüş Değerleri:

Fonksiyonlar, giriş parametreleri alabilir ve bir veya birden fazla değer döndürebilir.

```python
def topla(a, b):
    sonuc = a + b
    return sonuc

toplam = topla(3, 5)
print("Toplam:", toplam)
```

Bu örnekte, `topla` adlı bir fonksiyon, iki parametre alır ve bu parametrelerin toplamını döndürür.

## Varsayılan Parametreler:

Fonksiyonlara varsayılan parametreler atanabilir, böylece fonksiyon çağrılırken belirtilmeyen parametreler için varsayılan değerler kullanılır.

```python
def selam_ver(isim="Dünya"):
    print("Merhaba, " + isim + "!")

selam_ver("Ahmet")
selam_ver()
```

Bu örnekte, `selam_ver` adlı bir fonksiyon, bir isim parametresi alır ve eğer isim belirtilmemişse "Dünya" kullanılır.

## Esnek Parametreler:

Fonksiyonlar, esnek sayıda parametre alabilir. `*args` (pozisyonel argümanlar) ve `**kwargs` (anahtar kelime argümanları) kullanarak esnek parametreleri işleyebilirsiniz.

```python
def toplamlari_hesapla(*sayilar):
    toplam = sum(sayilar)
    return toplam

sonuc = toplamlari_hesapla(1, 2, 3, 4, 5)
print("Toplam:", sonuc)
```

Bu örnekte, `toplamlari_hesapla` adlı bir fonksiyon, esnek sayıda sayı alır ve bu sayıların toplamını döndürür.

## Lambda Fonksiyonları:

Lambda fonksiyonları, kısa süreli, isimsiz fonksiyonlardır. Genellikle tek bir satırda ifade edilebilecek küçük işlemler için kullanılır.

```python
kare_al = lambda x: x ** 2
print(kare_al(5))
```

Bu örnekte, `lambda` ile tanımlanan bir fonksiyon, bir sayının karesini alır.

Python'da fonksiyonlar, kodunuzu modüler hale getirerek tekrar kullanılabilir ve daha okunabilir kılar. Fonksiyonlar, büyük projelerde ve kod tabanlı uygulamalarda önemli bir yapı taşıdır.

Jupyter Notebook'u kullanmaya başlamadan önce,`pip install jupyter` Ardından, terminal veya komut istemcisine `jupyter notebook` komutunu yazarak Jupyter Notebook'u başlatabilirsiniz.

## 2. Yeni Bir Notebook Oluşturma

Jupyter Notebook'un ana ekranında "Yeni Not Defteri" düğmesine tıklayarak yeni bir Python 3 notebook oluşturalım.

![New Notebook](./Yeni_not_defteri.png)

Bu işlem sonucunda, yeni bir notebook açılacaktır.

## 3. Hücrelerle Çalışma

Her Jupyter Notebook, hücrelerden oluşur. İki temel hücre tipi vardır: "Code" (kod) ve "Markdown" (metin).

### Code Hücresi Örneği:

```python
# Bu bir Python kodu hücresidir
sayi = 5
kare = sayi ** 2
print(f"{sayi}'nin karesi: {kare}")
```

Bu kodu içeren hücreyi çalıştırmak için hücreye tıklayın ve Ctrl + Enter tuşlarına basın. Çıktı hemen altında görünecektir.

### Markdown Hücresi Örneği:

Markdown hücreleri metin ve açıklamalar eklemek için kullanılır. Aşağıdaki örnek, başlık ve liste içerir:

```markdown
# Başlık

Bu bir metin hücresidir. Aşağıda bir liste bulunmaktadır:

- Madde 1
- Madde 2
- Madde 3
```

Bu metin hücresini çalıştırdığınızda, markdown formatına uygun şekilde biçimlendirilmiş metni göreceksiniz.

## 4. Grafik ve Görselleştirme

Jupyter Notebook, verileri görselleştirmek için kullanışlı araçlar içerir. Örneğin, [Matplotlib](https://matplotlib.org/) kütüphanesini kullanarak bir grafik oluşturalım:

```python
import matplotlib.pyplot as plt

# Veri oluşturma
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Grafik oluşturma
plt.plot(x, y)
plt.xlabel('X Ekseni')
plt.ylabel('Y Ekseni')
plt.title('Basit Bir Grafik')
plt.show()
```

Bu kodu içeren hücreyi çalıştırdığınızda, grafiği göreceksiniz.
